```python
# Step 1: Import Necessary Libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Load the MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Step 3: Data Preprocessing
x_train = x_train / 255.0
x_test = x_test / 255.0

# Step 4: Build the Neural Network Model
model = keras.Sequential([
    # Flatten the 28x28 input images to a 1D array of 784 pixels
    keras.layers.Flatten(input_shape=(28, 28)),
    # Add a hidden layer with 128 neurons and ReLU activation function
    keras.layers.Dense(128, activation='relu'),
    # Add the output layer with 10 neurons (one for each digit) and softmax activation
    keras.layers.Dense(10, activation='softmax')
])

# Step 5: Compile the Model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Step 6: Train the Model
model.fit(x_train, y_train, epochs=5)

# Step 7: Evaluate the Model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {test_accuracy}')

# Step 8: Make Predictions
predictions = model.predict(x_test)

# Step 9: Visualize Results
for i in range(10):
    plt.imshow(x_test[i], cmap=plt.cm.binary)
    plt.xlabel(f'Actual: {y_test[i]} - Predicted: {np.argmax(predictions[i])}')
    plt.show()

# Step 10: Save and Deploy (Optional)
# You can save the trained model for later use or deployment in real applications.
model.save('mnist_digit_recognition_model.h5')
```
